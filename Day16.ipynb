## Day 16

https://adventofcode.com/2021/day/16

```python
from functools import reduce
from operator import mul

def hex2bits(h):
    b = bin(int(h,16))[2:]
    z = (4-len(b))*"0"
    return z+b

def string2bits(s):
    return "".join([ hex2bits(h) for h in s ])

def bits2values(B,vv=0,verbose=False,d=0):
    V = int(B[0:3],2)
    T = int(B[3:6],2)
    if T==4: # literal value
        if verbose: print(d*" "+"Literal  V:",V,end="")
        A = []
        i = 6
        while True:
            a = (int(B[i]))
            A.append(B[i+1:i+5])
            i+=5
            if a==0:
                break
        N = int("".join(A),2)
        if verbose: print("  N: {}".format(N))
        return vv+V,N,i   
    else: # operator
        I = int(B[6])
        if verbose: print(d*" "+"Operator V: {}  T: {}  I: {}".format(V,T,I),end="")
        vv += V
        iret = -1
        nn = []
        if I==0:
            L = int(B[7:7+15],2)
            if verbose: print("  L: {}".format(L))
            bsub = B[7+15:7+15+L]
            i = 0
            while i<L:
                vv,N,inew = bits2values(bsub[i:],vv,verbose,d+2)
                nn.append(N)
                i += inew
            iret = 7+15+L
        elif I==1:
            NP = int(B[7:7+11],2)
            if verbose: print("  NP: {}".format(NP))
            bsub = B[7+11:]
            i = 0
            np = 0
            while np<NP:
                vv,N,inew = bits2values(bsub[i:],vv,verbose,d+2)
                nn.append(N)
                i += inew
                np += 1
            iret = 7+11+i
        if T==0:
            N = sum(nn)
        elif T==1:
            N = reduce(mul,nn,1)
        elif T==2:
            N = min(nn)
        elif T==3:
            N = max(nn)
        elif T==5:
            if nn[0]>nn[1]:
                N = 1
            else:
                N = 0
        elif T==6:
            if nn[0]<nn[1]:
                N = 1
            else:
                N = 0        
        elif T==7:
            if nn[0]==nn[1]:
                N = 1
            else:
                N = 0 
        return vv,N,iret
```

### Part 1 tests

```python
def versionSum(s):
    print("*** {} ***".format(s))
    b = string2bits(s)
    V, N, _ = bits2values(b,0,verbose=True)
    print("Version sum: {}\n".format(V))

versionSum("D2FE28")
versionSum("38006F45291200")
versionSum("EE00D40C823060")
versionSum("8A004A801A8002F478")
versionSum("620080001611562C8802118E34")
versionSum("C0015000016115A2E0802F182340")
versionSum("A0016C880162017C3686B18A3D4780")
```
